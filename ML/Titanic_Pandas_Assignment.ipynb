```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
df.head()
```


## Question 1
How many passengers are there in the dataset?


```python
num_passengers = len(df)
print("Number of passengers:", num_passengers)
#OR we can use
#num_passengers = df.shape[0]
```


## Question 2
Find the number of male and female passengers.


```python
gender_counts = df['Sex'].value_counts()
print(gender_counts)
```


## Question 3
Calculate the average age of passengers (ignore missing values).


```python
avg_age = df['Age'].mean()
print("Average age of passengers:", avg_age)
```


## Question 4
What percentage of passengers survived?


```python
survival_rate = (df['Survived'].sum() / len(df)) * 100
print("Percentage of passengers who survived:", survival_rate)
```


## Question 5
Find survival rate by gender.


```python
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean() * 100
print(survival_rate_by_gender)
```


## Question 6 (Visualization)
Plot a bar chart showing survival rate by gender.


```python
import matplotlib.pyplot as plt
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean() * 100
plt.bar(survival_rate_by_gender.index, survival_rate_by_gender.values, color=['pink', 'lightblue'])
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate (%)")
plt.show()
```


## Question 7
Find survival rate by passenger class (Pclass).


```python
survival_rate_class = df.groupby('Pclass')['Survived'].mean() * 100
print(survival_rate_class)
```


## Question 8 (Visualization)
Plot a bar chart showing survival rate by passenger class.


```python
import matplotlib.pyplot as plt
survival_rate_class = df.groupby('Pclass')['Survived'].mean() * 100

plt.bar(survival_rate_class.index, survival_rate_class.values, color=['gold','silver', 'brown'])
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate (%)")
plt.xticks([1,2,3], ['First Class', 'Second Class', 'Third Class'])
plt.show()
```


## Question 9
Create a new column `Family_Size` using SibSp and Parch.


```python
# Create new column Family_Size
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1

# Display first few rows to check
df[['SibSp', 'Parch', 'Family_Size']].head()
```


## Question 10 (Visualization)
Plot a histogram of passenger ages.


```python
import matplotlib.pyplot as plt

plt.hist(df['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```
