```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
print(df.head())
```

### Q1. How many rows and columns are present in the Iris dataset?

```python
print(df.shape)
```

### Q2. How many unique species exist in the dataset?

```python
# Number of unique species
print(df['species'].nunique())

# Unique species labels
print(df['species'].unique())

# Actual name of species
species_name = iris.target_names
print(species_name)
```

### Q3. What is the average sepal length?

```python
avg_sepal_length = df['sepal length (cm)'].mean()
print(avg_sepal_length)
```

### Q4. How many flowers have a petal width > 1.5 cm?

```python
count = (df['petal width (cm)'] > 1.5).sum()
print(count)
```

### Q5. Are there any missing values in the dataset?

```python
print(df.isnull().sum())
```

### Q6. Which species has the largest average petal length?

```python
df['species'] = iris.target_names[df['species']]
avg_petal_length_by_species = df.groupby('species')['petal length (cm)'].mean()
print(avg_petal_length_by_species)
```

### Q7. Find the min, max, and mean of sepal width.

```python
print("MIN: ",df['sepal width (cm)'].min())
print("MAX: ",df['sepal width (cm)'].max())
print("MEAN: ",df['sepal width (cm)'].mean())
```

### Q8. How many samples have sepal_length > 6 AND petal_length > 5?

```python
count = ((df['sepal length (cm)'] > 6) & (df['petal length (cm)'] > 5)).sum()
print(count)
```

### Q9. Compute the correlation between sepal_length and petal_length.

```python
correlation = df['sepal length (cm)'].corr(df['petal length (cm)'])
print(correlation)
```

### Q10. For each species, find the median petal_width.

```python
df_median = df.groupby('species')['petal width (cm)'].median()
print(df_median)
```
